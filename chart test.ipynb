```python
from openpyxl import Workbook
from openpyxl.chart import BarChart, Series, Reference

wb = Workbook(write_only=True)
ws = wb.create_sheet()

rows = [
    ('Number', 'Batch 1', 'Batch 2'),
    (2, 10, 30),
    (3, 40, 60),
    (4, 50, 70),
    (5, 20, 10),
    (6, 10, 40),
    (7, 50, 30),
]


for row in rows:
    ws.append(row)


chart1 = BarChart()
chart1.type = "col"
chart1.style = 10
chart1.title = "Bar Chart"
chart1.y_axis.title = 'Test number'
chart1.x_axis.title = 'Sample length (mm)'

data = Reference(ws, min_col=2, min_row=1, max_row=7, max_col=3)
cats = Reference(ws, min_col=1, min_row=2, max_row=7)
chart1.add_data(data, titles_from_data=True)
chart1.set_categories(cats)
chart1.shape = 4
ws.add_chart(chart1, "A10")

# from copy import deepcopy

# chart2 = deepcopy(chart1)
# chart2.style = 11
# chart2.type = "bar"
# chart2.title = "Horizontal Bar Chart"

# ws.add_chart(chart2, "G10")


# chart3 = deepcopy(chart1)
# chart3.type = "col"
# chart3.style = 12
# chart3.grouping = "stacked"
# chart3.overlap = 100
# chart3.title = 'Stacked Chart'

# ws.add_chart(chart3, "A27")


# chart4 = deepcopy(chart1)
# chart4.type = "bar"
# chart4.style = 13
# chart4.grouping = "percentStacked"
# chart4.overlap = 100
# chart4.title = 'Percent Stacked Chart'

# ws.add_chart(chart4, "G27")

wb.save("bar.xlsx")
```

```python
#no title
# from openpyxl import Workbook
from openpyxl.chart import (
     RadarChart,
     Reference,
     Series,
 )

import openpyxl
from openpyxl.chart.axis import ChartLines
from openpyxl import drawing
wb = openpyxl.Workbook()
ws = wb.active

rows = [
    ['평가항목', '등급'],
    ['적용성', 8],
    ['친환경성', 9],
    ['시공성', 9],
    ['유지관리성', 9],
    ['품질관리성', 9],
    ['안정성', 6],
]

for row in rows:
    ws.append(row)

chart = RadarChart()
#chart.type = "filled"
chart.type = "standard" #"marker"#"standard" #line
labels = Reference(ws, min_col=1, min_row=2, max_row=7)
data = Reference(ws, min_col=1, max_col=2, min_row=2, max_row=7)
chart.add_data(data, titles_from_data=False)
chart.set_categories(labels)
chart.style = 2
chart.y_axis.majorGridlines = ChartLines()
lineProp = drawing.line.LineProperties(prstDash='dash')
#series.graphcalProperties.line =lineProp
chart.ChartLines = None
#chart.title = None
#chart.y_axis.majorGridlines = 1
chart.y_axis.ChartLines = lineProp
chart.legend = None

ws.add_chart(chart, "A8")

wb.save("radar.xlsx")
```
