```python
def numToExNum(i,j):
    col=[]
    if i==26:
        col.insert(0,"Z")
    else:        
        while i>0:
            col.insert(0,chr(i%26+64))
            i=i//26
    j=str(j)
    col.append(j)
    excol=''.join(col)
    return (excol)
```
